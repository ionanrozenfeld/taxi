```python
import pickle
import pandas as pd
```

```python
f=open('../network_mesh_space=0.0025_month=1_day=2.pickle','rb')
network = pickle.load(f)
f.close()
```

```python
count_pick_ups = network.loc(axis=0)[7,:,:,:]
count_drop_offs = network.loc(axis=0)[:,7,:,:]
```

```python
count_pick_ups_cols = count_pick_ups.reset_index()
count_drop_offs_cols = count_pick_ups.reset_index()
```

```python
pickups = pd.DataFrame(count_pick_ups_cols.groupby('pickup_center')[0].sum())
dropoffs = pd.DataFrame(count_drop_offs_cols.groupby('dropoff_center')[0].sum())
```

```python
pickups.head()
```

```python
dropoffs.head()
```

```python
new = pd.merge(pickups,dropoffs,how='outer',left_index=True, right_index=True)
```

```python
new.head()
```

```python
new = new.fillna(value=0)
```

```python
new = new.reset_index()
```

```python
new.head()
```

```python
map_data = pd.DataFrame()
```

```python
map_data['centers'] = new['index']
```

```python
map_data['activity'] = new['0_x'] + new['0_y']
```

```python
map_data['attractiveness'] = new['0_y'] - new['0_x']
```

```python
map_data['incoming'] = new['0_y']
```

```python
map_data['outgoing'] = new['0_x']
```

```python
map_data.head()
```

```python
map_data['centers'] = map_data['centers'].astype(int)
```

```python
f=open('../centers_long_lat_id_mesh_space=0.0025.pickle','rb')
grid = pickle.load(f)
f.close()
```

```python
grid.head()
```

```python
gridMap_data = pd.merge(map_data,grid,how='inner',left_on='centers',right_on='center_id')
```
